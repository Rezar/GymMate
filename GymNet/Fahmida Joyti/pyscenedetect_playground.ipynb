```python
video_manager.set_downscale_factor()
video_manager.start()
scene_manager.detect_scenes(frame_source=video_manager)
scene_list = scene_manager.get_scene_list(base_timecode)
```

```python
mp4_files = [x for x in os.listdir("./Youtube-CA-Workout Exercises") if x.endswith(".mp4")]
```

```python
for video in mp4_files:
    print(f"Splitting video {video}")
    file_path = 'Youtube-CA-Workout Exercises\\' + video
    video_manager = VideoManager([file_path])
    stats_manager = StatsManager()
    scene_manager = SceneManager(stats_manager)
    scene_manager.add_detector(ContentDetector())
    base_timecode = video_manager.get_base_timecode()
    video_manager.set_downscale_factor()
    video_manager.start()
    scene_manager.detect_scenes(frame_source=video_manager)
    scene_list = scene_manager.get_scene_list(base_timecode)
    scenedetect.video_splitter.split_video_ffmpeg(file_path, scene_list)
    video_manager.release()
```
