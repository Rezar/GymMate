```python
df_angle = pd.DataFrame(columns = ['name','angle_sort', 'delta_1','delta_2', 'delta_3',
                                  'delta_4', 'delta_5', 'delta_6', 'delta_7', 'delta_8'], index = df.index)
```

```python
for k in range(len(df)):
    X_ang = []
    A=[]
    a1_1 = min (eval (df.angle_1_list[k]))
    a2_1 = max (eval (df.angle_1_list[k]))
    delta_1 = a2_1-a1_1
    X_ang.append(delta_1)

    a1_2 = min (eval (df.angle_2_list[k]))
    a2_2 = max (eval (df.angle_2_list[k]))
    delta_2 = a2_2-a1_2
    X_ang.append(delta_2)

    a1_3 = min (eval (df.angle_3_list[k]))
    a2_3 = max (eval (df.angle_3_list[k]))
    delta_3 = a2_3-a1_3
    X_ang.append(delta_3)

    a1_4 = min (eval (df.angle_4_list[k]))
    a2_4 = max (eval (df.angle_4_list[k]))
    delta_4 = a2_4-a1_4
    X_ang.append(delta_4)

    a1_5 = min (eval (df.angle_5_list[k]))
    a2_5 = max (eval (df.angle_5_list[k]))
    delta_5 = a2_5-a1_5
    X_ang.append(delta_5)

    a1_6 = min (eval (df.angle_6_list[k]))
    a2_6 = max (eval (df.angle_6_list[k]))
    delta_6 = a2_6-a1_6
    X_ang.append(delta_6)

    a1_7 = min (eval (df.angle_7_list[k]))
    a2_7 = max (eval (df.angle_7_list[k]))
    delta_7 = a2_7-a1_7
    X_ang.append(delta_7)

    a1_8 = min (eval (df.angle_8_list[k]))
    a2_8 = max (eval (df.angle_8_list[k]))
    delta_8 = a2_8-a1_8
    X_ang.append(delta_8)
    arr = np.array(X_ang)
    sort_index = np.argsort(arr) 

    for i in range (8):
        if sort_index[i] == 0:
            A.append("delta_1")
        elif sort_index[i] == 1:
            A.append("delta_2")
        elif sort_index[i] == 2:
            A.append("delta_3")
        elif sort_index[i] == 3:
            A.append("delta_4")
        elif sort_index[i] == 4:
            A.append("delta_5")
        elif sort_index[i] == 5:
            A.append("delta_6")  
        elif sort_index[i] == 6:
            A.append("delta_7")
        elif sort_index[i] == 7:
            A.append("delta_8")
        else:
            pass
    df_angle.iloc[k]['name'] = df.Label[k]   
    df_angle.iloc[k]['angle_sort'] = A
    df_angle.iloc[k]['delta_1'] = delta_1
    df_angle.iloc[k]['delta_2' ]= delta_2
    df_angle.iloc[k]['delta_3'] = delta_3
    df_angle.iloc[k]['delta_4'] = delta_4
    df_angle.iloc[k]['delta_5'] = delta_5
    df_angle.iloc[k]['delta_6'] = delta_6
    df_angle.iloc[k]['delta_7'] = delta_7
    df_angle.iloc[k]['delta_8'] = delta_8
#df_angle.to_csv("Train.csv",index=False)
df_angle.to_csv("Test.csv",index=False)     
```
