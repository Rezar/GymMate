```python
import seaborn as sns
import pandas as pd
import numpy as np
dataset = pd.read_csv('AlphaPose.csv')
#dataset = pd.read_csv('OpenPose.csv')
#dataset = pd.read_csv('blazepose.csv')
```

```python
for i in range (len(dataset)):    
    t1=eval(dataset.X_RShoulder[i])
    t2=eval(dataset.Y_RShoulder[i])
    t3=eval(dataset.X_LShoulder[i])
    t4=eval(dataset.Y_LShoulder[i])
    t5=eval(dataset.X_RWrist[i])
    t6=eval(dataset.Y_RWrist[i])
    t7=eval(dataset.X_LWrist[i])
    t8=eval(dataset.Y_LWrist[i])
    t9=eval(dataset.X_RHip[i])
    t10=eval(dataset.Y_RHip[i])
    t11=eval(dataset.X_LHip[i])
    t12=eval(dataset.Y_LHip[i])
    t13=eval(dataset.X_RAnkle[i])
    t14=eval(dataset.Y_RAnkle[i])
    t15=eval(dataset.X_LAnkle[i])
    t16=eval(dataset.Y_LAnkle[i])
    t17=eval(dataset.X_Rknee[i])
    t18=eval(dataset.Y_Rknee[i])
    t19=eval(dataset.X_LKnee[i])
    t20=eval(dataset.Y_LKnee[i])
    a=[t1,t2,t3,t4,t5,t6,t7,t8,t9,t10,t11,t12,t13,t14,t15,t16,t17,t18,t19,t20]
    print(dataset.name[i])
    sns_plot = sns.heatmap(a,
                yticklabels=False,
                xticklabels=False,
                cbar=False,
               cmap="YlGnBu") 
    sns_plot.figure.savefig(f"plot/alphapose/{dataset.name[i]}.png")
```
