```python
import numpy as np
from numpy import array
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def connectpoints(x,y,z,p1,p2):
    x1, x2 = x[p1], x[p2]
    y1, y2 = y[p1], y[p2]
    z1, z2 = z[p1], z[p2]
    plt.plot([x1,x2],[y1,y2],[z1,z2],'k-')

def connectpoints_2d(x,y,p1,p2):
    x1, x2 = x[p1], x[p2]
    y1, y2 = y[p1], y[p2]
    plt.plot([x1,x2],[y1,y2],'k-')
```

```python
def rot_Y_ax(x,y,z,deg):
  theta = np.radians(deg)
  c, s = np.cos(theta), np.sin(theta)
  R = np.matrix([[c, 0, s], [0, 1,0], [-s,0,c]])
  v = np.matrix( [ x,  y, z] )
  A=(R*v.reshape(3,1))
  X_r,Y_r,Z_r=array(A).flat
  #print(X_r,Y_r,Z_r)
  return X_r,Y_r,Z_r

def rot_X_ax(x,y,z,deg):
  theta = np.radians(deg)
  c, s = np.cos(theta), np.sin(theta)
  R = np.matrix([[1, 0, 0], [0, c,-s], [0,s,c]])
  v = np.matrix( [ x,  y, z] )
  A=(R*v.reshape(3,1))
  X_r,Y_r,Z_r=array(A).flat
  #print(X_r,Y_r,Z_r)
  return X_r,Y_r,Z_r
```
