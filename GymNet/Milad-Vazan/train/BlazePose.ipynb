```python
from os import listdir
from os.path import isfile, join
import numpy
import cv2
import cv2
import numpy as np
import pandas as pd 
import mediapipe as mp
from google.colab.patches import cv2_imshow
```

```python
mypath='/content/drive/MyDrive/vf_frame/back extensions'
onlyfiles = [ f for f in listdir(mypath) if isfile(join(mypath,f)) ]
images = numpy.empty(len(onlyfiles), dtype=object)
x1_=[]
y1_=[]
 
x2_=[]
y2_=[]
 
x3_=[]
y3_=[]
 
x4_=[]
y4_=[]
 
x5_=[]
y5_=[]
 
x6_=[]
y6_=[]
 
x7_=[]
y7_=[]
 
x8_=[]
y8_=[]
 
x9_=[]
y9_=[]
 
 
x10_=[]
y10_=[]
 
for n in range(0, len(onlyfiles)):
  images[n] = join(mypath,onlyfiles[n])
  uploaded = images[n]
  name=uploaded
  images = {name: cv2.imread(name) }
  mp_pose = mp.solutions.pose
  mp_drawing = mp.solutions.drawing_utils 
  drawing_spec = mp_drawing.DrawingSpec(thickness=1, circle_radius=1)
  with mp_pose.Pose(
    static_image_mode=True, min_detection_confidence=0.5, model_complexity=2) as pose:
    for name, image in images.items():
      results = pose.process(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
      image_hight, image_width, _ = image.shape
      if not results.pose_landmarks:
        continue

  
      print(f'Pose landmarks of {name}:')
      x1_.append(results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_SHOULDER].x * image_width)
      y1_.append(results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_SHOULDER].y * image_hight)
    
      x2_.append(results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_SHOULDER].x * image_width)
      y2_.append(results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_SHOULDER].y * image_hight)
    
      x3_.append(results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_WRIST].x * image_width)
      y3_.append(results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_WRIST].y * image_hight)
    
      x4_.append(results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_WRIST].x * image_width)
      y4_.append(results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_WRIST].y * image_hight)
    
      x5_.append(results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_HIP].x * image_width)
      y5_.append(results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_HIP].y * image_hight)
    
      x6_.append(results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_HIP].x * image_width)
      y6_.append(results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_HIP].y * image_hight)
    
      x7_.append(results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_ANKLE].x * image_width)
      y7_.append(results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_ANKLE].y * image_hight)
    
      x8_.append(results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_ANKLE].x * image_width)
      y8_.append(results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_ANKLE].y * image_hight)

      x9_.append(results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_KNEE].x * image_width)
    
      y9_.append(results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_KNEE].y * image_hight)
    
      x10_.append(results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_KNEE].x * image_width)
      y10_.append(results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_KNEE].y * image_hight)
    
df_x = pd.DataFrame({"name":mypath,'RIGHT SHOULDER X':[x1_], 'RIGHT SHOULDER Y':[y1_],
                      'LEFT SHOULDER X':[x2_], 'LEFT SHOULDER Y':[y2_],
                      'RIGHT WRIST X':[x3_], 'RIGHT WRIST Y':[y3_],
                      'LEFT WRIST X':[x4_], 'LEFT WRIST Y':[y4_],
                      'RIGHT HIP X':[x5_], 'RIGHT HIP Y':[y5_],
                      'LEFT HIP X':[x6_], 'LEFT HIP Y':[y6_],
                      'RIGHT ANKLE X':[x7_], 'RIGHT ANKLE Y':[y7_],
                      'LEFT ANKLE X':[x8_], 'LEFT ANKLE Y':[y8_],
                      'RIGHT KNEE X':[x9_], 'RIGHT KNEE Y':[y9_],
                      'LEFT KNEE X':[x10_], 'LEFT KNEE Y':[y10_]
})
df_x.to_csv("/content/sample_data/m8.csv",index=False)
```
